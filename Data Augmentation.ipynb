# Import Dependencies

```python
from niacin.text import en
from niacin.augment import RandAugment
```

# Create an augmentor which can do synonym replacement and random swap

```python
augmentor=RandAugment([
    en.add_synonyms,
    en.add_hypernyms,
    en.swap_words
],
n=2,
m=50
)
```

# Data Preprocessing

```python
import pandas as pd
df = pd.read_csv("data.csv")
df.head()
```

# Test the augmentor for first message

```python
for tx in augmentor:
    data=tx(df['Message'][0])
```

```python
data
```

# Create a new data frame for augmented data

```python
data = {'Category':[],'Message':[]}  
  
# Create DataFrame  
df2 = pd.DataFrame(data)  
  
# Print the output.  
print(df2)
```

# Append the augmented data into new data frame

```python
for i in range(1000):
    for tx in augmentor:
        new_message=tx(df['Message'][i])
        new_category=df['Category'][i]
        new_row = {'Category':new_category, 'Message':new_message}
        df2 = df2.append(new_row, ignore_index=True)
```
